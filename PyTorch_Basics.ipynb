###**TENSOR DATA TYPES**
**Note:** PyTorch dtypes are one of the three big errors in PyTorch and DL

1.Tensor not the right Data Type

2.Tensor not the right shape

3.Tensor not on the right device


```python
#float32 data type
float_32_tensor = torch.tensor( [3.,4.,5.],
                               dtype=None,#What Data type is the tensor(e.g float32,float16)
                               device=None,#What device your Tensor is on
                               requires_grad=False)#Whether or not to track gradients with this tensors operations
float_32_tensor
```

```python
float_16_tensor = float_32_tensor.type(torch.float16)
float_16_tensor.dtype
float_16_tensor.device
int_32_tensor = torch.tensor([3,4,5],dtype = torch.int32)
int_32_tensor.device
```

***manipulating tensors***

Tensor Operations include

* Addition

* Subtraction

* Multiplication(Element wise)

* Division

* Matrix-Multiplication


```python
#Tensor operations(Modifying tensors)
#Adding
tensor = torch.tensor([1,2,3])
tensor + 10 #Add 10
```

```python
# multiply tensor by 10
mult_tensor = tensor * 10
```

```python
#Dividing tensors by 2
mult_tensor/2
```

```python
#subtract ten
mult_tensor - 10
```

**In-Built functions of PyTorch**

```python
#In-Built functions of PyTorch Multiplication
torch.mul(tensor,2)
```

```python
#Addition
torch.add(tensor,5)
```

```python
#Subtraction
torch.sub(tensor,1)
```

```python
#Division
torch.div(mult_tensor,2)
```
