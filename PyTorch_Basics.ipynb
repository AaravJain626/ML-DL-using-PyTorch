##Reshaping, stacking, squeezing and unsqueezing tensors

* Reshaping - reshapes and input tensor to desired shape

* View - Returns view of an input tensor of certain shape but keep same memory as the original tensor

* Stacking - combines multiple tensors on top of each other (vstack) or side-by-side (hstack)

* Squeez - removes all the `1` dimension from the tensor

* unsqueez - ass a `1` dimension to a target tensor

* permute - return the view of the input with dimensions permuted(swapped) in certain way



```python
#tensor
x = torch.arange(1.,11.)
x,x.shape
```

```python
reshaped_tensor = x.reshape(2,5)
reshaped_tensor,reshaped_tensor.shape
```

```python
#change the view
z = x.view(2,5)
z, z.size()
```

```python
#changing in z will change in `x` as it shares memeory with the original one
```

```python
#stacking tensors on top of each other
x_stacked = torch.stack([x,x,x,x], dim=0)
x_stacked
```

```python
#torch.squeez removes all single dimensions form a target tensor
test_tensor = torch.tensor([[1,4,1,3,1,4,1,5]])
test_tensor.shape
test_tensor.squeeze().shape
```

```python
tensor_x = torch.tensor([[[1,2,4,6,2,5]]])
tensor_x.shape
```

```python
tensor_x.squeeze().shape
squeezed_tensor_x = tensor_x.squeeze()
```

```python
#torch.permute -rearranges the dimensions of a target tensor in a specified order
x_original = torch.rand(size=(224,224,3)) #[Height,width,color_channels]
#permute the origibal trensor to rearrange the axes or dims of the order
x_permuted = x_original.permute(2,0,1) #shifts axes 0->1,1->2,2->0
print(f"previous shape: {x_original.shape}")
print(f"New shape: {x_permuted.shape}")#[Color_chanel,height,width]
```

## Reproducbility(Trying to take random out of random)

neural N working

`Start with randon no.->tensor operations -> update random numbers to try and make them better representation of data ->again-->again->again...`

To reduce the randomness in pyTorch and NNs there is a concept of **Random Seed**.

what the random does is 'flavour' the randomness


```python
random_tensor_A = torch.rand(3,4)
random_tensor_B = torch.rand(3,4)
random_tensor_A, random_tensor_B
```

```python
RANDOM_SEED = 42
torch.manual_seed(RANDOM_SEED)
random_tensor_C = torch.rand(3,4)

torch.manual_seed(RANDOM_SEED)
random_tensor_D = torch.rand(3,4)
#torch.manual_seed
print(random_tensor_C, random_tensor_D)
print(random_tensor_C == random_tensor_D)
```
